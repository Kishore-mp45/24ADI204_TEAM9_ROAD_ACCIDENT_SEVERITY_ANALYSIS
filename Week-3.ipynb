IQR Outlier Filtering

### Outlier Detection and Removal Using IQR Method

```python
def iqr_filter(df, col):
    Q1 = df_1[col].quantile(0.25)
    Q3 = df_1[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return df_1[(df[col] >= lower) & (df_1[col] <= upper)]
```

### Checking the Shape of the Dataset

```python
iqr_cols = [
    'Distance(mi)',
    'Visibility(mi)',
    'Precipitation(in)',
    'Humidity(%)'
]

for col in iqr_cols:
    df_1 = iqr_filter(df_1, col)

print("After IQR filtering:", df_1.shape)
```

###  Temporal Heatmap — Accidents by Hour of Day vs. Day of Week
This heatmap extracts temporal features from the `Start_Time` column to reveal accident frequency patterns across different times of day and days of the week.

```python
import pandas as pd

# Parse Start_Time and extract temporal features
df_1['Start_Time_Parsed'] = pd.to_datetime(df_1['Start_Time'], errors='coerce')
df_1['Hour'] = df_1['Start_Time_Parsed'].dt.hour
df_1['DayOfWeek'] = df_1['Start_Time_Parsed'].dt.day_name()

# Build pivot table
temp_df = df_1.dropna(subset=['Hour', 'DayOfWeek'])
pivot = temp_df.groupby(['DayOfWeek', 'Hour']).size().unstack(fill_value=0)
days_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
pivot = pivot.reindex(days_order)

plt.figure(figsize=(16, 6))
sns.heatmap(pivot, cmap='YlOrRd', linewidths=0.3, linecolor='white')
plt.title('Accident Frequency: Hour of Day vs. Day of Week', fontsize=16, fontweight='bold')
plt.xlabel('Hour of Day (0–23)', fontsize=12)
plt.ylabel('Day of Week', fontsize=12)
plt.tight_layout()
plt.show()

# Clean up helper columns
df_1.drop(columns=['Start_Time_Parsed', 'Hour', 'DayOfWeek'], inplace=True, errors='ignore')
```

### Feature Engineering — Skewness Correction

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95, random_state=42)
X_pca = pca.fit_transform(X_scaled)
```

### Checking the Shape of the Dataset

```python
print("Original shape:", X.shape)
print("Reduced shape after PCA:", X_pca.shape)
print("Explained Variance Ratio:")
print(pca.explained_variance_ratio_)
print("Total Variance Retained:", sum(pca.explained_variance_ratio_))
```

```python
# Cumulative variance
cum_var = np.cumsum(pca.explained_variance_ratio_)

plt.figure(figsize=(8, 5))
plt.plot(cum_var, marker='o')
plt.xlabel("Number of Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("PCA - Variance Explained")
plt.grid()
plt.show()
```

**Interpretation :**
- PCA (Principal Component Analysis) was applied to reduce the dimensionality of the feature space while retaining maximum variance.
- The explained variance ratio plot shows the contribution of each principal component to the total variance.
- A threshold of 95% cumulative variance was used to determine the optimal number of components to retain.
- This dimensionality reduction step accelerates model training and helps mitigate the curse of dimensionality.

## Weather Condition vs. Accident Severity (100% Stacked Bar Chart)

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# 1. Decode Weather_Condition back to original categorical strings
weather_decoded = label_encoders['Weather_Condition'].inverse_transform(df_1['Weather_Condition'])

# 2. Create a temporary DataFrame
temp_df = pd.DataFrame({
    'Weather': weather_decoded,
    'Severity': df_1['Severity']
})

# 3. Identify Top 5 weather conditions
top_weather = temp_df['Weather'].value_counts().head(5).index
filtered_df = temp_df[temp_df['Weather'].isin(top_weather)]

# 4. Calculate 100% stacked proportions using cross-tabulation
crosstab = pd.crosstab(filtered_df['Weather'], filtered_df['Severity'], normalize='index') * 100

# 5. Plot
ax = crosstab.plot(kind='bar', stacked=True, figsize=(10, 6), colormap='viridis', edgecolor='black', alpha=0.8)
plt.title('Proportional Severity of Accidents by Top 5 Weather Conditions', fontsize=14, fontweight='bold')
plt.xlabel('Weather Condition', fontsize=12)
plt.ylabel('Percentage of Accidents (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.legend(title='Severity', bbox_to_anchor=(1.05, 1), loc='upper left')

# Annotate the bars with percentages
for c in ax.containers:
    ax.bar_label(c, label_type='center', fmt='%.1f%%', size=9, color='white', weight='bold')

plt.tight_layout()
plt.show()
```
